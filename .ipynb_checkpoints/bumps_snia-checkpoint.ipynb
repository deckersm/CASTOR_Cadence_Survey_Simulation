```python
def temperature(t):
    a = 10**13
    k = 0.2
    return 2.5e4 * (a**0.25) * (k**(-35./36.)) * (t**(-37./72.))
```

```python
def blackbody(lamb, T):
    K = constants.k
    h = constants.h
    c = constants.c
    return ((2*h*c**2)/(lamb**5)) * (1./(np.exp(h*c/lamb*K*T) - 1))
```

```python
lamb = np.arange(1000, 10000, 100)*1e-10
b = blackbody(lamb*1e-10, 2e8)
plt.plot(lamb, b)
```

```python
h = 6.626e-34
c = 3.0e+8
k = 1.38e-23

def planck(lamb, T):
    a = 2.0*h*c**2
    b = h*c/(lamb*k*T)
    return a/ ( (lamb**5) * (np.exp(b) - 1.0) )


# generate x-axis in increments from 1nm to 3 micrometer in 1 nm increments
# starting at 1 nm to avoid wav = 0, which would result in division by zero.
wavelengths = np.arange(1e-9, 3e-6, 1e-9) 

# intensity at 4000K, 5000K, 6000K, 7000K
intensity4000 = planck(wavelengths, 4000.)
intensity5000 = planck(wavelengths, 5000.)
intensity6000 = planck(wavelengths, 6000.)
intensity7000 = planck(wavelengths, 7000.)


plt.plot(wavelengths*1e10, intensity4000, 'r-') 
# plot intensity4000 versus wavelength in nm as a red line
plt.plot(wavelengths*1e10, intensity5000, 'g-') # 5000K green line
plt.plot(wavelengths*1e10, intensity6000, 'b-') # 6000K blue line
plt.plot(wavelengths*1e10, intensity7000, 'k-') # 7000K black line

plt.xlabel(r'Wavelength [$\AA$]')
```

```python
wavelengths = np.arange(1e-9, 5e-7, 1e-9) 
fig = plt.figure(dpi=150, facecolor = 'w', figsize = (8,6))
ax = fig.add_subplot(111)

for t in [temps[1]]:
    flux = planck(wavelengths, t/7.3e3)
    ax.plot(wavelengths*1e10, flux) 
ax.set_xlabel(r'Wavelength [$\AA$]')
ax.axvline(x=1000)
```

```python
# Core collapse rates from Strolger (2015)

def f(z):
    fx = np.piecewise(z, [(z <= 0.08) * (z > 0), 
                        (z <= 0.29) * (z > 0.08), 
                        (z <= 0.47) * (z > 0.29), 
                        (z <= 0.72) * (z > 0.29), 
                        (z <= 1.56) * (z > 0.72), 
                        (z > 1.56)], 
                        [0.72e-4, 1.33e-4, 1.81e-4, 3.91e-4, 3.22e-4, 3.76e-4])
    return fx

x = np.arange(0,2, 0.01)
rate = f(x)

plt.plot(x, rate)
plt.yscale('log')

x1 = [0.04, 0.25, 0.38, 0.59, 1.14, 1.93]
y = [0.72e-4, 1.33e-4, 1.81e-4, 3.91e-4, 3.22e-4, 3.76e-4]

plt.scatter(x1, y)
```
