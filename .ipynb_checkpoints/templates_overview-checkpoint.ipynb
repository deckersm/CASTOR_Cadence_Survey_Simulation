```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import glob
import math
```

```python
spectra = pd.DataFrame(columns = ['type', 'phase', 'lamb', 'flux'])
```

```python
# importing SN Ib/c plasticc templates

f = glob.glob('/Users/maximedeckers/Documents/RSE/CASTOR/Templates/NON1ASED.SNIbc-Templates/*.SED')
t = 'snibc'
for filename in f:
    df = pd.DataFrame(pd.read_csv(filename, sep='\s+', names = ['phase', 'lamb', 'flux']))

    phases = list(set(df['phase']))
    
    m = filename.split('/')[-1].split('-')[1].replace('.SED', '')
    
    for p in phases:
        df_ = df.loc[df['phase']==p]
        lamb = list(df_['lamb'])
        flux = list(df_['flux'])
        spectra.loc[len(spectra), ['type', 'phase', 'lamb', 'flux']] = '91bg', p, lamb, flux

        d_new = pd.DataFrame(data={'#lamb':lamb, 'flux':flux})
        d_new.to_csv('individual_spectral_templates/snibc/SED_snibc_{}_{}d.dat'.format(m, p), index = False, encoding='utf-8', sep = ' ')
```

```python
# importing Hsiao

df = pd.DataFrame(pd.read_csv('/Users/maximedeckers/Documents/RSE/CASTOR/Templates/hsiao_template/snflux_1a.dat', sep='\s+', names = ['phase', 'lamb', 'flux']))
print(df)
phases = list(set(df['phase']))
print(phases)
t = 'Hsiao'
for p in phases:
    if int(p) < 10:
        df_ = df.loc[df['phase']==p]
        lamb = list(df_['lamb'])
        flux = list(df_['flux']* 5e7)
        spectra.loc[len(spectra), ['type', 'phase', 'lamb', 'flux']] = 'hsiao', p, lamb, flux
        
        d_new = pd.DataFrame(data={'#lamb':lamb, 'flux':flux})
        d_new.to_csv('individual_spectral_templates/snia/SED_snia_hsiao_{}d.dat'.format(p), index = False, encoding='utf-8', sep = ' ')
```
