```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
min_lam_castor = 1500
min_lam_plasticc = 1000

redshift_array = np.arange(0, 3, 0.2)

min_lam_plasticc_obs = []
for z in redshift_array:
    min_lam_plasticc_obs.append(min_lam_plasticc * (1. + z))
```

```python
plt.figure(dpi=150, figsize = (8, 6), facecolor = 'w')
plt.plot(redshift_array, min_lam_plasticc_obs)
plt.xlabel('Redshift')
plt.ylabel('Minimum wavelength of observed PLASTICC spectrum')
plt.axhline(y=1500, linestyle = '--', color = 'red')
plt.axvline(x=0.5, linestyle = '--', color = 'red')

plt.show()
```
