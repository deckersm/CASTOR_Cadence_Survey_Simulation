# Notebook to produce range of Hsiao models with varying peak mags and widths to match observed distributions

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sncosmo
import math
import glob
from numpy import random
import sncosmo
```

```python
# Function that takes input peak magnitude and outputs the stretch value
def stretch(mb_stretched, mb = -19.25, alpha = 1.52):
    return ((mb_stretched - mb) / alpha ) + 1

```

```python
# importing CASTOR passbands to sncosmo

for band in ['uv', 'u', 'g']:
    df = pd.DataFrame(pd.read_csv('/Users/maximedeckers/Documents/PhD/Photometry/Filters/passband_castor.{}'.format(band), sep='\s+', names = ['wavelength', 'transmission']))
    df.loc[:, 'wavelength'] = df['wavelength']*10000
    if band == 'uv':
        df = df.loc[(df['wavelength']<3500) & (df['wavelength']>1100)]
    elif band == 'u':
        df = df.loc[(df['wavelength']<4500) & (df['wavelength']>1100)]
    elif band == 'g':
        df = df.loc[(df['wavelength']<6000) & (df['wavelength']>1100)]
        
    wavelength = df['wavelength']
    transmission = df['transmission']
    band_data = sncosmo.Bandpass(wavelength, transmission, name=band)
    sncosmo.register(band_data, band, force = True)
```

```python
# Creating normal distribution of absolute peak magnitudes for SNe Ia
# Using values from Richardson (2014)

mean = -19.26
std = 0.51

peak_mags = random.normal(loc=mean, scale=std, size=1000)
peak_spec = 'individual_spectral_templates/snia/SED_snia_hsiao_0.0d.dat'
factors = np.arange(0, 4, 0.01)
multiplication_factors = []
count = -1
for mb in peak_mags:
    count+=1
    df = pd.DataFrame(pd.read_csv(peak_spec, names = ['lamb', 'flux'], comment = '#', sep = '\s+'))
    spectrum = sncosmo.Spectrum(df['lamb'], df['flux'])
    mag = 0
    for f in factors:
        print(len(multiplication_factors), count)
        spectrum = sncosmo.Spectrum(df['lamb'], df['flux'] * f)
        mag = spectrum.bandmag(['g'], magsys='ab')[0]
        if np.round(mag, 1) == np.round(mb, 1):
            multiplication_factors.append(f)
            break
        #else:
        #    print(np.round(mag, 1), np.round(mb, 1))


print(len(multiplication_factors))
```

```python
# Applying the stretch to the Hsiao template
# Multiply the time axis by the stretch (s)

# importing Hsiao

df = pd.DataFrame(pd.read_csv('/Users/maximedeckers/Documents/RSE/CASTOR/Templates/hsiao_template/snflux_1a.dat', sep='\s+', names = ['phase', 'lamb', 'flux']))
phases = list(set(df['phase']))

t = 'Hsiao'

for i in range(len(peak_mags)):
    mb = peak_mags[i]
    factor = multiplication_factors[i]
    for p in phases:
        df_ = df.loc[df['phase']==p]
        s = stretch(mb)
        phase = p * s
        lamb = list(df_['lamb'])
        flux = list(df_['flux']* 5e7*factor)
        #spectra.loc[len(spectra), ['type', 'phase', 'lamb', 'flux']] = 'hsiao', phase, lamb, flux
        
        d_new = pd.DataFrame(data={'#lamb':lamb, 'flux':flux})
        d_new.to_csv('individual_spectral_templates/snia/SED_snia_peakmag{}_{}d.dat'.format(np.round(mb, 2), np.round(phase, 2)), index = False, encoding='utf-8', sep = ' ')
```

```python
specs = glob.glob('individual_spectral_templates/snia/SED_snia_peakmag*')
fig = plt.figure(dpi=150, figsize = (8,6), facecolor = 'w')
ax = fig.add_subplot(111)

for i in range(len(peak_mags)):
    mb = peak_mags[i]
    factor = multiplication_factors[i]
    phot = pd.DataFrame(columns = ['phase', 'filter', 'abs_mag'])
    for s in specs:
        if str(np.round(mb, 2)) in s:
            phase = float(s.split('/')[-1].split('_')[3].replace('d.dat', ''))
            df = pd.DataFrame(pd.read_csv(s, names = ['lamb', 'flux'], comment = '#', sep = '\s+'))
            spectrum = sncosmo.Spectrum(df['lamb'], df['flux']*factor)
            
            for band in ['g']:
                mag = spectrum.bandmag([band], magsys='ab')[0]
                phot.loc[len(phot), ['phase', 'filter', 'abs_mag']] = phase, band,mag

    phot = phot.loc[phot['abs_mag']<30]
    
    for band in ['g']:
        ax.scatter(phot.loc[phot['filter']==band, 'phase'], phot.loc[phot['filter']==band, 'abs_mag'], label = band)
    
ax.invert_yaxis()
ax.set_xlabel('Phase [d]')
ax.set_ylabel('AB absolute magnitude')
plt.show()
plt.close()
```
